# Extração de Dados em Big Data ETL com Apache Spark

A primeira parte do processo de ETL é a extração de dados dos sistemas de origem, sejam esses sistemas CRMs, ERPs, logs de máquinas, etc. As fontes variam de acordo com o negócio, empresa e principalmente o contexto dos dados a serem armazenados. Uma vez que os dados são extraídos de seus sistemas principais, eles devem ser armazenados em um repositório intermediário, ficando disponível para a manipulação da próxima etapa.

Veja abaixo exemplos das fontes de dados mais utilizadas:

- Arquivos de Texto/CSV;
- Excel;
- Banco de Dados Relacionais;
- JSON;
- Redes Sociais;
- Imagens;
- Vídeos;
- Páginas Web.

![etl](https://uaismart.com/wp-content/uploads/2018/02/o-que-e-ETL.png)

Este tutorial apresenta como é possível fazer a extração de dados de algumas fontes de dados utilizando o Apache Spark.

## Fontes de Dados

O Apache Spark suporta a conexão com uma variedade de fontes de dados. Este tutorial fornece uma descrição geral e links para todas as diferentes fontes de dados suportadas pelo Apache Spark. As fontes de dados CSV, Parquet e JSON serão abordadas de forma detalhada neste tutorial por não precisarem de integração com bancos de dados ou outras formas mais complexas de fontes de dados e serem mais fáceis de replicar no ambiente local fora da plataforma Databricks.

[As principais fontes de dados disponíveis para o Apache Spark](https://docs.databricks.com/spark/latest/data-sources/index.html) estão listadas abaixo:

- [JDBC](https://docs.databricks.com/spark/latest/data-sources/sql-databases.html): você pode consultar muitos bancos de dados SQL utilizando drivers JDBC.
- [Amazon Redshift] (): você pode carregar dados em um Spark SQL DataFrame a partir do Amazon Redshift e escrever de volta para tabelas do Redshift. Esta fonte de dados é mais indicada para ETL do que consultas interativas, devido ao grande volume de dados extraídos do S3.
- [Amazon S3](https://docs.databricks.com/spark/latest/data-sources/aws/amazon-s3.html): você pode ler e escrever dados no [Amazon S3](https://aws.amazon.com/s3/), um serviço para armazenar um grande volume de dados não estruturados.
- [Azure Blob Storage](https://docs.databricks.com/spark/latest/data-sources/azure/azure-storage.html): outra fonte de dados é o [Azure Blob Storage](https://azure.microsoft.com/en-us/services/storage/blobs/) que permite lidar um grande volume de dados não estruturados.
- [Azure Data Lake Storage](https://docs.databricks.com/spark/latest/data-sources/azure/azure-datalake-gen2.html): você pode integrar sua solução Spark com o [Data Lake da Azure](https://docs.microsoft.com/en-us/azure/storage/data-lake-storage/introduction) para Big Data Analytics.
- [Cosmos DB](https://docs.databricks.com/spark/latest/data-sources/azure/cosmosdb-connector.html): você pode conectar sua solução com este [banco de dados distribuído da Microsoft](https://azure.microsoft.com/en-us/services/cosmos-db/) que permite que você escale sua solução em clusters de computadores.
- [Azure SQL Data Warehouse](https://docs.databricks.com/spark/latest/data-sources/azure/sql-data-warehouse.html): existe suporte para integração com o [Azure SQL Data Warehouse](https://azure.microsoft.com/en-us/services/sql-data-warehouse/), que é um data warehouse na nuvem que permite executar consultas complexas sobre petabytes de dados de forma paralela. 
- [Cassandra](https://docs.databricks.com/spark/latest/data-sources/cassandra.html): existe um [conector da DataStax](https://github.com/datastax/spark-cassandra-connector) que permite a integração do Apache Spark com o Apache Cassandra.
- [Couchbase](https://docs.databricks.com/spark/latest/data-sources/couchbase.html): o Couchbase mantém um [conector com o Spark](https://docs.couchbase.com/spark-connector/2.2/index.html) que permite de forma fácil a integração de ambos.
- [ElasticSearch](https://docs.databricks.com/spark/latest/data-sources/elasticsearch.html): o Elasticsearch possui um [integrador](https://www.elastic.co/guide/en/elasticsearch/hadoop/current/spark.html) com o Apache Spark que mapeia boa parte das funcionalidades do Elasticsearch para o Apache Spark.
- [Hive](https://docs.databricks.com/spark/latest/data-sources/importing-hive-tables.html): Spark possui integração nativa com o Hive.
- [MongoDB](https://docs.databricks.com/spark/latest/data-sources/mongodb.html): o MongoDB é um banco de dados orientado a documentos que armazena os dados no formato JSON e o [integrador](https://docs.mongodb.com/spark-connector/current/) com o Spark é mantido pela equipe do MongoDB.
- [Neo4j](https://docs.databricks.com/spark/latest/data-sources/neo4j.html): o Neo4j é um banco de dados de grafos que permite integrar com o Spark através de APIs para o RDD, DataFrame, GraphX e GraphFrames. Esta integração é feita com um [conector](https://github.com/neo4j-contrib/neo4j-spark-connector) que permite transferir dados entre o servidor Neo4j e o Spark.
- [Redis](https://docs.databricks.com/spark/latest/data-sources/redis.html): O Redis é um armazenamento de estrutura de dados de chave-valor na memória. Ele mantém um [pacote](http://spark-packages.org/package/RedisLabs/spark-redis) de integração com o Spark.
- [Riak Time Series](https://docs.databricks.com/spark/latest/data-sources/riak-ts.html): este banco de dados é otimizado para IoT e séries temporais e possui um [integrador](https://spark-packages.org/package/basho/spark-riak-connector) com o Spark mantido pela equipe do Riak.
- [Snowflake](https://docs.databricks.com/spark/latest/data-sources/snowflake.html): o Snowflake é um data warehouse para cloud com integração com diversas fontes de dados. A Databricks e o Snowflake desenvolveram juntos um [integrador](https://docs.snowflake.net/manuals/user-guide/spark-connector-use.html#) com o Spark.
- [Arquivos Avro](https://docs.databricks.com/spark/latest/data-sources/read-avro.html): o Apache Avro é um sistema de serialização de dados com suporte a várias linguagens. O Apache Spark é capaz de ler e escrever arquivos no formato Avro de forma nativa através de uma biblioteca embutida.
- [Arquivos Orc](https://spark.apache.org/docs/latest/sql-data-sources-orc.html): o Spark possui também suporte a arquivos no formato ORC (Optimized Row Columnar).
- [Arquivos XML](https://github.com/databricks/spark-xml): o Spark possui suporte a XML utilizando um plugin escrito pela Databricks.

### Leitura de Imagens pelo Apache Spark

A partir da versão 2.4 do Apache Spark ficou muito mais fácil carregar arquivos de [imagens](https://spark.apache.org/docs/latest/ml-datasource) (jpeg, png, etc.) a partir de um diretório. Esta nova funcionalidade veio para facilitar a construção dos modelos de Aprendizado de Máquina com os dados de imagens. No Spark é criado um DataFrame com uma única coluna dentro do StructType, denominada "image", contendo o esquema padrão da imagem:

- origin: StringType (caminho para o arquivo da imagem)
- height: IntegerType (altura da imagem)
- width: IntegerType (largura da imagem)
- nChannels: IntegerType (número de canais da imagem. Geralmente com valor 1 para escala de cinza, 3 para coloridas e 4 para coloridas com o canal alpha)
- mode: IntegerType (fornece informação (compatível com OpenCV) de como interpretar o campo)
- data: BinaryType (dado armazenado no formato binário)

Neste tutorial iremos carregar um conjunto de imagens do [Projeto CAVIAR](http://groups.inf.ed.ac.uk/vision/CAVIAR/CAVIARDATA1/) com pessoas andando, descansando, fazendo compras, etc. Este dataset está armazenado na plataforma Databricks e foi escolhido porque você pode evoluir sua solução com imagens para identificar atitudes suspeitas utilizando Deep Learning com [sparkdl](https://github.com/databricks/spark-deep-learning) dentro da plataforma Databricks, tendo como referência o tutorial deste [blog post](https://databricks.com/blog/2018/09/13/identify-suspicious-behavior-in-video-with-databricks-runtime-for-machine-learning.html).

### Leitura do CSV

O Spark tem uma biblioteca nativa para leitura e escrita de arquivos no formato CSV. Nesta Seção iremos apresentar alguns exemplos da leitura de arquivos no formato CSV com tratamento de erros para arquivos mal formatados. 

As principais opções para leitura dos arquivos CSV pelo Spark são listadas abaixo. O restante das opções pode ser vista na [documentação de leitura do CSV no Spark](https://spark.apache.org/docs/latest/api/python/pyspark.sql.html?highlight=read%20csv#pyspark.sql.DataFrameReader.csv).

- **path**: caminho dos arquivos ou diretório de arquivos CSV.
- **header**: valor *true* para usar a primeira linha do arquivo como cabeçalho (default: *false*).
- **sep**: caractere delimitador das colunas (default: *,*). 
- **schema**: você pode informar o esquema do arquivo CSV a ser lido utilizando o *pyspark.sql.types.StructType* com uma string formatada como DDL (por exemplo: col0 INT, col1 DOUBLE).
- **quote**: pode ser configurado com qualquer caractere para delimitar um texto. Delimitadores dentro das *quotes* serão ignoradas (default: *"*).
- **escape**: pode ser configurado como qualquer caractere para escape no texto. Caracteres *quote* serão ignorados (default: *\*).
- **mode**: indica a forma do parsing do CSV. Por padrão é PERMISSIVE. Os valores possíveis são:
  - PERMISSIVE: tenta fazer o parsing de todas as linhas: valores nulls são inseridos em colunas vazias e colunas extras são ignoradas.
  - DROPMALFORMED: exclui as linhas que tem menos ou mais colunas do que o esperado pelo esquema.
  - FAILFAST: aborta a leitura do arquivo com uma exceção se alguma linha mal formatada for encontrada.
- **encoding**: codificação do arquivo (default: *UTF-8*).
- **inferSchema**: valor *true* para automaticamente inferir o tipo de cada coluna (default: *false*).
- **comment**: você pode inserir comentários dentro do arquivo. Este parâmetro indica qual caractere será utilizado como comentário e ignora as linhas que começam com este caractere. É possível desabilitar comentários no arquivo configurando este valor como *null*. (default: *#*).
- **dateFormat**: string que indica o formato da data seguindo os padrões definidos pelo [java.text.SimpleDateFormat](https://docs.oracle.com/javase/7/docs/api/java/text/SimpleDateFormat.html). O formato padrão é 'yyyy-MM-dd'.

Os exemplos abaixo usam o dataset *diamonds* disponível no [diretório de datasets da Databricks](https://docs.databricks.com/getting-started/databricks-datasets.html#databricks-datasets).

```
%fs
head /databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv
```

Agora iremos ler o arquivo *diamonds.csv* tendo a primeira linha como cabeçalho e indicando para inferir o esquema automaticamente. Modifique os parâmetros de *header* e *inferSchema* para false para ver o que acontece!

```
# leitura do arquivo e atribuição para o Dataframe diamonds
diamonds = spark.read.csv('/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv', header='true', inferSchema='true')
# apresenta os dados do csv na tela no formato tabular
display(diamonds)
```

```
diamonds.printSchema()
```

Você também pode especificar o esquema do CSV com o parâmetro *schema*. No exemplo abaixo, definimos o esquema do CSV dando nomes as colunas e o tipo de cada uma. Repare na coluna *index* do tipo String que antes se chamava *_c0* e era do tipo Integer. Quando o esquema é definido, as opções *header* e *inferSchema* são ignoradas.

```
from pyspark.sql.types import *
schema = StructType([StructField("index", StringType(), True),
                     StructField("carat", DoubleType(), True),
                     StructField("cut", StringType(), True),
                     StructField("color", StringType(), True),
                     StructField("clarity", StringType(), True),
                     StructField("depth", DoubleType(), True),
                     StructField("table", DoubleType(), True),
                     StructField("price", IntegerType(), True),
                     StructField("x", DoubleType(), True),
                     StructField("y", DoubleType(), True),
                     StructField("z", DoubleType(), True)])
diamonds = spark.read.csv('/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv', header='false', schema=schema)
display(diamonds)
```

```
diamonds.printSchema()
```

Na leitura dos arquivos CSV com um esquema definido pelo usuário, é possível que os dados não atendam ao esquema especificado. Os próximos exemplos mostram diferentes tratamentos de erros dependendo do modo escolhido pelo usuário: PERMISSIVE, DROPMALFORMED ou FAILFAST.

```
#modo PERMISSIVE

from pyspark.sql.types import *
schema = StructType([StructField("index", StringType(), True),
                     StructField("carat", DoubleType(), True),
                     StructField("cut", StringType(), True),
                     StructField("color", StringType(), True),
                     StructField("clarity", StringType(), True),
                     StructField("depth", IntegerType(), True),  #depth deveria ser do tipo Double
                     StructField("table", DoubleType(), True),
                     StructField("price", IntegerType(), True),
                     StructField("x", DoubleType(), True),
                     StructField("y", DoubleType(), True),
                     StructField("z", DoubleType(), True),
                     StructField("_corrupt_record", StringType(), True)])
diamonds_permissive = spark.read.csv('/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv', header='true', inferSchema='true', schema=schema)
display(diamonds_permissive)
```

```
# seleciona as linhas que atendem ao esquema definido
goodRows = diamonds_permissive.filter(diamonds_permissive._corrupt_record.isNotNull())
display(goodRows)
```

```
# seleciona as linhas que atendem ao esquema definido
badRows = diamonds_permissive.filter(diamonds_permissive._corrupt_record.isNotNull())
display(badRows)
```

```
# mode DROPMALFORMED. Para o modo FAILFAST basta trocar DROPMALFORMED por FAILFAST dentro de options. 
diamonds_malformed = spark.read.csv('/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv', header='true', inferSchema='true', mode='DROPMALFORMED', schema=schema)
display(diamonds_malformed)
```

### Leitura de Arquivos Parquet

O [Parquet](http://parquet.io/) é um formato de arquivo de armazenamento colunar e está disponível para qualquer projeto no ecossistema Hadoop, independentemente da escolha da estrutura de processamento de dados, modelo de dados ou linguagem de programação. O Apache Spark possui suporte nativo para arquivos no formato Parquet. O formato colunar permite ler, descompactar e processar apenas as colunas necessárias para a consulta atual tornando ele mais eficiente que formatos baseados em linhas como o CSV, por exemplo. Parquet é otimizado para trabalhar com dados complexos utilizando diferentes formas de compressão de dados e tipos de codificação. Ele é extremamente indicado para aquelas consultas que precisam ler um conjunto de colunas de uma grande tabela, já que o Parquet lê apenas as colunas necessárias minimizando o impacto do I/O. Veja na figura abaixo a [comparação de desempenho entre o Parquet e CSV](https://blog.openbridge.com/how-to-be-a-hero-with-powerful-parquet-google-and-amazon-f2ae0f35ee04).

![parquet](https://cdn-images-1.medium.com/max/2400/1*Tmw-dLB3m74Uv-cCCTph2Q.png)

No exemplo apresentado abaixo, primeiramente a tabela é escrita no disco com os atributos dos dispositivos IoT. Depois deste passo, arquivo é lido no formato Parquet do disco utilizando o Spark.

```
devices_df.write.mode("overwrite").parquet("/tmp/testParquet")
devices_parquet = spark.read.parquet("/tmp/testParquet")
display(devices_parquet)
```

É possível visualizar o arquivo Parquet escrito no disco que foi utilizado na leitura. Os detalhes na escrita do arquivo no formato Parquet serão detalhados em outro tutorial.

```
%sh
ls /dbfs/tmp/testParquet
```

Você também pode ler o arquivo Parquet armazenado em vários diretórios com diferentes esquemas, onde o Spark fica responsável pelo merge automático do esquema com a opção 'mergeSchema' igual a *true*.

```
from pyspark.sql import Row

# Cria o primeiro dataframe com as colunas 'single' e 'double'
squaresDF = spark.createDataFrame(sc.parallelize(range(1, 6))
                                  .map(lambda i: Row(single=i, double=i ** 2)))
squaresDF.write.mode("overwrite").parquet("tmp/test_table/key=1")

# Cria outro dataframe no mesmo diretório 'tmp/test_table' com um esquema diferente, contendo as colunas 'single' e 'triple'
cubesDF = spark.createDataFrame(sc.parallelize(range(1, 6))
                                .map(lambda i: Row(single=i, triple=i ** 3)))
cubesDF.write.mode("overwrite").parquet("tmp/test_table/key=2")

# Lê o diretório pai com as duas tabelas particionadas passando a opção 'mergeSchema' para fazer um merge dos esquemas das tabelas
mergedDF = spark.read.option("mergeSchema", "true").parquet("tmp/test_table")
mergedDF.printSchema()
display(mergedDF)
```
